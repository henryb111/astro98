```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt",
skiprows=1)

np.shape(data) #2642 rows, 2 columns
```

```python
magnitude = data[:,0]
temperature = data[:,1]

print("max magnitude", np.max(magnitude))
print("corresponding to", np.argmax(magnitude))
print("min magnitude", np.min(magnitude))
print("corresponding to", np.argmin(magnitude))
print("max temperature", np.max(temperature))
print("corresponding to", np.argmax(temperature))
print("min tempearture", np.min(temperature))
print("corresponding to", np.argmin(temperature))
```

```python
#3.3 Plot and HR Diagrams

log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

fig, ax = plt.subplots(figsize=(8, 6))

sc = ax.scatter(log_temperature, log_magnitude, c=log_temperature, cmap='viridis_r', s=10)

cb = plt.colorbar(sc, orientation='horizontal', label='Log temperature')     
cb.ax.invert_xaxis()
plt.scatter(log_temperature,log_magnitude)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.scatter(log_temperature[264],log_magnitude[264], color = 'black', marker = '*', s = 300, label = 'brightest')
plt.scatter(log_temperature[1988],log_magnitude[1988], color = 'purple', marker = '*', s = 300, label = 'dimmest')
plt.scatter(log_temperature[2613],log_magnitude[2613], color = 'green', marker = '*', s = 300, label = 'hottest')
plt.scatter(log_temperature[1923],log_magnitude[1923], color = 'orange', marker = '*', s = 300, label = 'coldest')

plt.title('Hertzsprung-Russel Diagram')
plt.xlabel('Temperature (log K)')
plt.ylabel('Magnitude')
plt.legend(loc='best')

plt.show()
```
