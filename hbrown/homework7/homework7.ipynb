```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2 Matplotlib

def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
    
x = np.linspace(0, 2 * np.pi, 1000)

amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

fig, axis = plt.subplots(1, 3, figsize=(25, 5))

for A, w in zip(amplitude, frequency):

    y1 = make_sine_wave(x, A, 1)
    axis[0].plot(x, y1)
    
    y2 = make_sine_wave(x, 1, w)
    axis[1].plot(x, y2)
    
    y3 = make_sine_wave(x, A, w)
    axis[2].plot(x, y3)

plt.tight_layout()
plt.show()
```
