```python
#4
arr = np.array([
[2, 3, 5],
[14, 6, 8],
[11, 13, 17],
[7, 10, 13]
])

def isPrime(n):
    if n < 2
        return False
    if n == 2:
        return True
    for i in range(2,np.sqrt(n)+1)
        if n % i == 0:
            return False
    return True
    
```
